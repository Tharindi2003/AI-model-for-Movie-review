```python
# Import scikit-learn tools
# CountVectorizer will turn our text into numbers the AI can understand.
# MultinomialNB is the actual AI model algorithm we will use.
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
x_train =[
     "The movie was amazing and fantastic",
    "I really liked the acting",
    "A great plot with excellent scenes",
    "This was a terrible and boring movie",
    "I did not like the film at all",
    "The acting was awful"
]

y_train = [1,1,1,0,0,0]
```

```python
ectorize = CountVectorizer()

x_train_vectorize = vectorize.fit_transform(x_train)
```

```python
vectorize = CountVectorizer()

x_train_vectorize = vectorize.fit_transform(x_train)
```

```python
print(x_train_vectorize.toarray())
```

```python
model = MultinomialNB()
model.fit(x_train_vectorize, y_train)
```

```python
x_test =[
    "The acting was really great",
    "That was a horrible movie"
]

x_test_vectorize = vectorize.transform(x_test)

predictions = model.predict(x_test_vectorize)

print("Predictions for our new review: ")
for review,prediction in zip(x_test,predictions):
   if prediction == 1:
      print(f"'{review}' -> Positive")
   else:
      print(f"'{review}' -> Negative")
```
